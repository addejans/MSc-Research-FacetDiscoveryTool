```python
import numpy as np
import pandas as pd

file=open('atLeast2535.txt')
counter = 0
count2 = 0

data = [[]for i in range(42)]
for word in file.read().lower().split():
    if word not in ['x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8', '<=']:
        data[count2].append(word)
        counter += 1
        if counter%6 == 0:
            count2 += 1
           
RHS = []
for i in range(42):
    if data[i][5] not in RHS:
        RHS.append(data[i][5])
        
groupRHS = {}

for e in RHS:
    for i in range(42):
        if data[i][5] == e:
            if e not in groupRHS:
                groupRHS[e] = []
            groupRHS[e].append(data[i])

            
groupRHS
d=groupRHS
from pandas import Series
pd.DataFrame(dict([ (k,Series(v)) for k,v in d.iteritems() ]))
```
