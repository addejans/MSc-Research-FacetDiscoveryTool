## All Different Generation

```python
import random
from random import shuffle, Random
 
alldifferent = []
templst = []
temp2 = []

def minj(num, dim):
    mini = 0
    for i in range(len(dim)):
        if dim[i] == 1:
            mini += (i)
            num -= 1
    if num > 0:        
        dim.append(mini)
        dim.append(1)
        alldifferent.append(dim)
        
def maxk(num, dim):
    maxi = 0
    for i in dim:
        if i == 1:
            maxi += num
            num -= 1
    if num > 0:
        dim.append(maxi)
        dim.append(2)
        alldifferent.append(dim)
            
def listset(dimension):
    templst[:] = []
    temp = [0 for _ in range(dimension)]
    for i in range(len(temp)):
        temp[i] = 1
        temp2 = temp[:]
        templst.append(temp2)

        
#dimension = 5
for dimension in range(4,7):
    for k in range(40,dimension-1,-1):
        listset(dimension)
        for i in templst:
            minj(k, i)

    for k in range(40,dimension-1,-1):
        listset(dimension)
        for i in templst:
            maxk(k, i)

for e in alldifferent:
    if e[-1] == 1:
        for i in range(len(e)-2):
            e[i] *= -1
            

seed = 10        
random.Random(seed).shuffle(alldifferent)
alldifferent
```
